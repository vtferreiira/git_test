##### 1. Elabore um programa em Python que insere a tabuada de multiplicação de 9 em um arquivo txt.

```python
#Nome do arquivo
nome_arquivo = "tabuadaDe9.txt"

#Abre/Cria o arquivo em modo de escrita
with open(nome_arquivo, "w") as arquivo:
    #Loop de 1 a 10
    for i in range(1, 11):
        resultado = 9 * i
        linha = f"9 x {i} = {resultado}\n"

        #Escreve no arquivo
        arquivo.write(linha)

print("Tabuada do 9 salva com sucesso no arquivo!")

with open("tabuadaDe9.txt", "r") as arquivo:
    print("\nTabuada de 9:\n")
    print(arquivo.read())
```

##### 4. Elabore um programa em Python que leia o nome e duas notas de um aluno do teclado, calcule a média e armazene em um arquivo txt o nome, a média final e se o mesmo foi Aprovado (média >=6) ou Reprovado (média < 6)

```python
# Exercício 4 - Média e Status do Aluno
nome = input("Nome do aluno: ")
n1 = float(input("Nota 1: "))
n2 = float(input("Nota 2: "))
media = (n1 + n2) / 2
condicao = "Aprovado" if media >= 6 else "Reprovado"

with open("condicao_aluno.txt", "a", encoding="utf-8") as arquivo:
    arquivo.write(f"\nAluno: {nome}\nMédia Final: {media:.2f}\nStatus: {condicao}")
print("Resultado salvo em 'condicao_aluno.txt'.")
```

##### 5. Faça um algoritmo em Python que leia dois números inteiros do teclado, faça uma mini calculadora (soma, subtração, multiplicação e divisão) e armazene todos os resultados em um arquivo txt

```python
# Exercício 5 - Mini Calculadora
n1 = int(input("Digite o primeiro número inteiro: "))
n2 = int(input("Digite o segundo número inteiro: "))

operacoes = {
    "Soma": n1 + n2,
    "Subtração": n1 - n2,
    "Multiplicação": n1 * n2,
    "Divisão": n1 / n2 if n2 != 0 else "Erro (Divisão por zero)"
}

with open("calculadora.txt", "w", encoding="utf-8") as arquivo:
    for op, res in operacoes.items():
        arquivo.write(f"{op}: {res}\n")
print("Resultados salvos em 'calculadora.txt'.")
```

##### 1. (slide 63) Elabore um programa em Python que leia uma cadeia de DNA e gera a cadeia inversa. Faça a leitura da cadeia utilizando arquivo txt. Exemplo: Entrada: AATCTGCAC Saída: CACGTCTAA

```python
dna_usuario = input("Digite a sequência de DNA: ")

with open('dna.txt', 'w') as arquivo:
    arquivo.write(dna_usuario)

with open('dna.txt', 'r') as arquivo:
    conteudo = arquivo.read().strip()

inversa = conteudo[::-1]

print("-" * 20)
print("Entrada (Digitada):", conteudo)
print("Saída (Inversa):   ", inversa)
```

##### 2. (slide 63) Faça um programa que leia um texto de um arquivo txt e conte quantas palavras tem nesse arquivo, sem considerar os espaços

```python
nome_arquivo = "texto.txt"

with open(nome_arquivo, 'w', encoding='utf-8') as arquivo:
    texto = input("Digite um texto: ")
    arquivo.write(texto)

with open(nome_arquivo, 'r', encoding='utf-8') as arquivo:
    conteudo = arquivo.read()

palavras = conteudo.split()
quantidade = len(palavras)

print("O arquivo contém", quantidade, "palavras.")
```
